```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

## Beef Price Data

```python
# Load up all beef price datasets
df_all_uncooked_ground = pd.read_csv("https://raw.githubusercontent.com/milazz27/Data-Science-Project/main/Beef_Prices/avg_price_all_uncooked_ground.csv")
df_ground_beef = pd.read_csv("https://raw.githubusercontent.com/milazz27/Data-Science-Project/main/Beef_Prices/avg_price_ground_beef.csv")
df_beef_steaks = pd.read_csv("https://raw.githubusercontent.com/milazz27/Data-Science-Project/main/Beef_Prices/avg_price_beef_steaks.csv")
df_steak_sirloin = pd.read_csv("https://raw.githubusercontent.com/milazz27/Data-Science-Project/main/Beef_Prices/avg_price_steak_sirloin_choice_boneless.csv")
df_ground_chuck = pd.read_csv("https://raw.githubusercontent.com/milazz27/Data-Science-Project/main/Beef_Prices/avg_price_ground_chuck.csv")
df_ground_lean = pd.read_csv("https://raw.githubusercontent.com/milazz27/Data-Science-Project/main/Beef_Prices/avg_price_ground_lean.csv")
df_chuck_roast = pd.read_csv("https://raw.githubusercontent.com/milazz27/Data-Science-Project/main/Beef_Prices/avg_price_chuck_roast.csv")
df_price_steak = pd.read_csv("https://raw.githubusercontent.com/milazz27/Data-Science-Project/main/Beef_Prices/avg_price_steak_round.csv")
df_beef_roast = pd.read_csv("https://raw.githubusercontent.com/milazz27/Data-Science-Project/main/Beef_Prices/avg_price_beef_roast.csv")
df_round_roast = pd.read_csv("https://raw.githubusercontent.com/milazz27/Data-Science-Project/main/Beef_Prices/avg_price_round_roast.csv")
df_beef_stew = pd.read_csv("https://raw.githubusercontent.com/milazz27/Data-Science-Project/main/Beef_Prices/avg_price_beef_stew.csv")
```

```python
# Changing price avg column name to be more descriptive
df_all_uncooked_ground = df_all_uncooked_ground.rename(columns={'APU0000FC1101': 'avg_price_all_uncooked_ground'}) #
df_ground_beef = df_ground_beef.rename(columns={'APU0000703112': 'avg_price_ground_beef'}) #
df_beef_steaks = df_beef_steaks.rename(columns={'APU0000FC3101': 'avg_price_beef_steaks'}) #
df_steak_sirloin = df_steak_sirloin.rename(columns={'APU0000703613': 'avg_price_steak_sirloin'}) #
df_ground_chuck = df_ground_chuck.rename(columns={'APU0000703111': 'avg_price_ground_chuck'}) #
df_ground_lean = df_ground_lean.rename(columns={'APU0000703113': 'avg_price_ground_lean'}) #
df_chuck_roast = df_chuck_roast.rename(columns={'APU0000703213': 'avg_price_chuck_roast'}) #
df_price_steak = df_price_steak.rename(columns={'APU0000703511': 'avg_price_steak'}) #
df_beef_roast = df_beef_roast.rename(columns={'APU0000FC2101': 'avg_price_beef_roast'}) #
df_round_roast = df_round_roast.rename(columns={'APU0000703311': 'avg_price_round_roast'}) #
df_beef_stew = df_beef_stew.rename(columns={'APU0000703432': 'avg_price_beef_stew'}) #
```

```python
# From prior observation, earliest datapoint is from 1980 (recorded in ground chuck and beef stew dfs)

# Going to start joining these
df_1 = df_beef_stew.merge(df_ground_chuck, on='observation_date', how='left')
df_2 = df_1.merge(df_all_uncooked_ground, on='observation_date', how='left')
df_3 = df_2.merge(df_ground_beef, on='observation_date', how='left')
df_4 = df_3.merge(df_beef_steaks, on='observation_date', how='left')
df_5 = df_4.merge(df_steak_sirloin, on='observation_date', how='left')
df_6 = df_5.merge(df_ground_lean, on='observation_date', how='left')
df_7 = df_6.merge(df_chuck_roast, on='observation_date', how='left')
df_8 = df_7.merge(df_price_steak, on='observation_date', how='left')
df_9 = df_8.merge(df_beef_roast, on='observation_date', how='left')
df_all_avg_beef_prices = df_9.merge(df_round_roast, on='observation_date', how='left')

df_all_avg_beef_prices
```

```python
df_all_avg_beef_prices.to_csv('all_avg_beef_prices.csv', index=False)
```

```python
# Convert 'observation_date' to datetime objects if not already
df_all_avg_beef_prices['observation_date'] = pd.to_datetime(df_all_avg_beef_prices['observation_date'])

# Identify price-related columns (excluding 'observation_date')
price_columns = [col for col in df_all_avg_beef_prices.columns if 'avg_price' in col]

plt.figure(figsize=(15, 8))

for column in price_columns:
    sns.lineplot(data=df_all_avg_beef_prices, x='observation_date', y=column, label=column)

plt.title('Average Beef Prices Over Time for Various Cuts')
plt.xlabel('Observation Date')
plt.ylabel('Average Price Per Pound')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Feed Data


```python
def parse_century(date_str):
    dt = pd.to_datetime(date_str, format='%m/%d/%y')
    year = dt.year
    if year > 2026:
        dt = dt.replace(year=year - 100)
    return dt
```

```python
# Cleaning Corn Futures Data

df_corn_futures = pd.read_csv("https://raw.githubusercontent.com/milazz27/Data-Science-Project/main/Feed_Prices/corn_futures.csv")

# Grab only desired columns
df_corn_futures = df_corn_futures.drop(columns=['Change', 'Bid', 'Ask', 'Open Interest', '% Change', 'Last'])
df_corn_futures

# Convert to monthly by averaging
df_corn_futures['Date'] = df_corn_futures['Date'].apply(parse_century)
df_corn_futures

# Filter out rows where both attributes are missing
df_corn_futures = df_corn_futures[~(df_corn_futures['Settlement Price'].isna() & df_corn_futures['CVol'].isna())]
df_corn_futures

# Convert to Monthly Format
df_monthly_corn_futures = (
    df_corn_futures.set_index('Date')
      .resample('MS') # Changed 'ME' to 'MS' to get month start
      .mean(numeric_only=True)
      .reset_index()
)

df_monthly_corn_futures = df_monthly_corn_futures.rename(columns={'Date':'observation_date'})
df_monthly_corn_futures

df_monthly_corn_futures['observation_date'] = pd.to_datetime(df_monthly_corn_futures['observation_date'])
df_monthly_corn_futures
```

```python
# loading and cleaning ppi feed data
df_ppi_feed = pd.read_csv("https://raw.githubusercontent.com/milazz27/Data-Science-Project/main/Feed_Prices/ppi_feed.csv")
df_ppi_feed

df_ppi_feed = df_ppi_feed.rename(columns={'PCU311119311119H':'index_dec_2011'})
df_ppi_feed

df_ppi_feed['observation_date'] = pd.to_datetime(df_ppi_feed['observation_date'])
df_ppi_feed
```

```python
df_monthly_corn_futures
```

```python
# renaming to prevent collision
df_monthly_corn_futures = df_monthly_corn_futures.rename(columns={'Settlement Price':'corn_futures_settlement_price'})
df_monthly_corn_futures = df_monthly_corn_futures.rename(columns={'CVol':'corn_futures_CVol'})
df_monthly_corn_futures

# joining feed ppi with feed prices
df_feed_data = df_monthly_corn_futures.merge(df_ppi_feed, on='observation_date', how='left')
df_feed_data
```

```python
# Merging with comprehensive df
all_data = all_data.merge(df_feed_data, on='observation_date', how='left')
all_data
```

## Economic Data


```python
# Loading in feeder cattle futures
df_feeder_cattle_futures = pd.read_csv('https://raw.githubusercontent.com/milazz27/Data-Science-Project/main/Economic_Data/feeder_cattle_futures.csv')
df_feeder_cattle_futures

# Drop irrelevant cols
df_feeder_cattle_futures = df_feeder_cattle_futures.drop(columns=['Change', 'Bid', 'Ask', 'Open Interest', '% Change', 'Last'])
df_feeder_cattle_futures

# Convert to monthly by averaging
df_feeder_cattle_futures['Date'] = df_feeder_cattle_futures['Date'].apply(parse_century)


# Filter out rows where both attributes are missing
df_feeder_cattle_futures = df_feeder_cattle_futures[~(df_feeder_cattle_futures['Settlement Price'].isna() & df_feeder_cattle_futures['CVol'].isna())]

# Convert to Monthly Format
df_feeder_cattle_futures = (
    df_feeder_cattle_futures.set_index('Date')
      .resample('MS') # Changed 'ME' to 'MS' to get month start
      .mean(numeric_only=True)
      .reset_index()
)

df_feeder_cattle_futures = df_feeder_cattle_futures.rename(columns={'Date':'observation_date'})
df_feeder_cattle_futures

```

```python
# Live cattle futures

df_live_cattle_futures = pd.read_csv('https://raw.githubusercontent.com/milazz27/Data-Science-Project/main/Economic_Data/live_cattle_futures.csv')
df_live_cattle_futures

# Drop irrelevant columns
df_live_cattle_futures = df_live_cattle_futures.drop(columns=['Change', 'Bid', 'Ask', 'Open Interest', '% Change', 'Last'])
df_live_cattle_futures

# Convert to monthly by averaging
df_live_cattle_futures['Date'] = df_live_cattle_futures['Date'].apply(parse_century)
df_live_cattle_futures

# Convert to Monthly Format
df_live_cattle_futures = (
    df_live_cattle_futures.set_index('Date')
      .resample('MS') # Changed 'ME' to 'MS' to get month start
      .mean(numeric_only=True)
      .reset_index()
)

df_live_cattle_futures

df_live_cattle_futures = df_live_cattle_futures.rename(columns={'Date':'observation_date'})
df_live_cattle_futures
```

```python
df_live_cattle_futures
```

```python
# Merging cattle futures data
df_live_cattle_futures = df_live_cattle_futures.rename(columns={'Settlement Price':'live_cattle_futures_settlement_price'})
df_live_cattle_futures = df_live_cattle_futures.rename(columns={'CVol':'live_cattle_futures_CVol'})
df_live_cattle_futures

df_feeder_cattle_futures = df_feeder_cattle_futures.rename(columns={'Settlement Price':'feeder_cattle_futures_settlement_price'})
df_feeder_cattle_futures = df_feeder_cattle_futures.rename(columns={'CVol':'feeder_cattle_futures_CVol'})
df_feeder_cattle_futures

# Renaming to avoid collison
df_cattle_futures = df_live_cattle_futures.merge(df_feeder_cattle_futures, on='observation_date', how='left')
df_cattle_futures
```

```python
# merge cattle futures into all_data
all_data = all_data.merge(df_cattle_futures, on='observation_date', how='left')
all_data
```

```python
# Loading CPI data
df_cpi = pd.read_csv('https://raw.githubusercontent.com/milazz27/Data-Science-Project/main/Economic_Data/cpi_us.csv')
df_cpi
```

```python
# Melt the DataFrame to transform date columns into rows
df_cpi_melted = df_cpi.melt(id_vars=['Unnamed: 0'], var_name='Date', value_name='cpi')

# Convert 'Date' column to datetime objects
df_cpi_melted['Date'] = pd.to_datetime(df_cpi_melted['Date'], format='%m/%Y')

# Convert 'cpi' column to numeric, coercing errors to NaN
df_cpi_melted['cpi'] = pd.to_numeric(df_cpi_melted['cpi'], errors='coerce')

# Extract the year from the 'Date' column
df_cpi_melted['Year'] = df_cpi_melted['Date'].dt.year

# Group by year and calculate the mean CPI for each year
df_cpi_annual = df_cpi_melted.groupby('Year')['cpi'].mean().reset_index()

# Keep 'Year' column for merging
df_cpi_annual = df_cpi_annual[['Year', 'cpi']]

display(df_cpi_annual.head())
display(df_cpi_annual.tail())
```
